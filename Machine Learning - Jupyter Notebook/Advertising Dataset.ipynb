# Regress√£o Linear
## Least Squares Method

```python
import pandas as pd

publi = pd.read_csv('https://raw.githubusercontent.com/XMUR/ISLR-1/master/Advertising.csv', index_col=0)
publi.head()
```

```python
print(publi.shape)
```

```python
x = publi[['TV', 'Radio', 'Newspaper']]
y = publi['Sales']
```

```python
x.shape
```

```python
y.shape
```

```python
import seaborn as sns
sns.pairplot(publi, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height = 7, kind='reg')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

## Importa√ß√£o do modelo de regress√£o linear

```python
from sklearn.linear_model import LinearRegression
reglin = LinearRegression()
reglin.fit(x_train, y_train)
```

## Coeficiente: Aumento nas vendas para cada 1un  investida em publicidade

```python
list(zip(['TV', 'Radio', 'Newspaper'], reglin.coef_))
```

## Previs√£o

```python
y_prev = reglin.predict(x_test)
print(y_prev)
```

# Avalia√ß√£o da performance
## MAE(Mean absolute error)

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test, y_prev))
```

## MSE (Mean squared error)

```python
print(metrics.mean_squared_error(y_test, y_prev))
```

# RMSE (Root mean squared error)

```python
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test,y_prev)))
```
